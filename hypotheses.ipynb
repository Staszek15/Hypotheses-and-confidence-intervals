# TESTY DLA MU

```python
def test_1(alpha,hipo):
    k = 1000
    sigma=0.2
    mu = 1.5
    N=1000
    if hipo == "ne":
        row = 1-alpha/2
        quantile = scipy.stats.norm.ppf(row,0,1)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Z_test = (np.mean(X_vector,axis=1)-mu)/sigma*np.sqrt(k)
        answer=  (Z_test <= -quantile) + (Z_test >= quantile)
        return np.mean(answer)
    if hipo == "gt":
        row = 1-alpha
        quantile = scipy.stats.norm.ppf(row,0,1)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Z_test = (np.mean(X_vector,axis=1)-mu)/sigma*np.sqrt(k)
        answer= (Z_test >= quantile)
        return np.mean(answer)
    if hipo == "lt":
        row = 1-alpha
        quantile = scipy.stats.norm.ppf(row,0,1)
        X_vector = scipy.stats.norm.rvs(mu,sigma,(N,k))
        Z_test = (np.mean(X_vector,axis=1)-mu)/sigma*np.sqrt(k)
        answer= (Z_test <= -quantile)
        return np.mean(answer)
```

# BŁĘDY II RODZAJU

```python
def second_mistake_zad1(mu,alternative = 'lt',test_mu = 1.5,alpha = 0.05, n=1000):
    sigma = 0.2
    sum = 0
    if alternative == 'lt':
        #H1: mu < 1.5 
        for _ in range(1000):
            X = scipy.stats.norm.rvs(mu,sigma,n)
            Z = (np.mean(X)  - test_mu)/ sigma * np.sqrt(n)
            if Z >= scipy.stats.norm.ppf(alpha):
                sum += 1
    elif alternative == 'gt':
        #H1: mu > 1.5 
        for _ in range(1000):
            X = scipy.stats.norm.rvs(mu,sigma,n)
            Z = (np.mean(X)  - test_mu)/ sigma * np.sqrt(n)
            if Z <= scipy.stats.norm.ppf(1-alpha):
                sum += 1
    elif alternative == 'ne':
        #H1: mu != 1.5
        for _ in range(1000):
            X = scipy.stats.norm.rvs(mu,sigma,n)
            Z = (np.mean(X)  - test_mu)/ sigma * np.sqrt(n)
            if Z <= scipy.stats.norm.ppf(1-alpha/2) and Z >= scipy.stats.norm.ppf(alpha/2):
                sum += 1
    return sum/1000
```

```python
def second_mistake_zad2(test_sigma2,alternative = 'lt',alpha = 0.05, n=1000):
    mu = 0.2
    sigma2 = 1.5
    sum = 0
    if alternative == 'lt':
        #H1: sigma ^2 < 1.5 
        for _ in range(1000):
            X = scipy.stats.norm.rvs(mu,np.sqrt(test_sigma2),n)
            X2 = (n-1)*np.var(X)/sigma2
            if X2 >= scipy.stats.chi2.ppf(alpha,n-1):
                sum += 1
    elif alternative == 'gt':
        #H1: sigma ^2 > 1.5 
        for _ in range(1000):
            X = scipy.stats.norm.rvs(mu,np.sqrt(test_sigma2),n)
            X2 = (n-1)*np.var(X)/sigma2
            if X2 <= scipy.stats.chi2.ppf(1-alpha,n-1):
                sum += 1
    elif alternative == 'ne':
        #H1: sigma ^2 != 1.5
        for _ in range(1000):
            X = scipy.stats.norm.rvs(mu,np.sqrt(test_sigma2),n)
            X2 = (n-1)*np.var(X)/sigma2
            if X2 >= scipy.stats.chi2.ppf(alpha/2,n-1) and X2 <= scipy.stats.chi2.ppf(1-alpha/2,n-1):
                sum += 1
    return sum/1000
```
